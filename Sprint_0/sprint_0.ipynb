```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
import streamlit as st
```

```python
#Ejemplo numpy

# Crear un array NumPy
array1 = np.array([1, 2, 3, 4, 5])

# Crear un array 2D (matriz)
matriz = np.array([[1, 2, 3], [4, 5, 6]])

# Operaciones matemáticas elementales
array_sum = array1 + 10     # Sumar 10 a cada elemento
array_squared = array1 ** 2 # Elevar cada elemento al cuadrado

# Operaciones entre arrays
array2 = np.array([10, 20, 30, 40, 50])
array_add = array1 + array2 # Suma elemento a elemento

# Calcular la media y la suma
mean_array1 = np.mean(array1)  # Media del array
sum_array1 = np.sum(array1)    # Suma de los elementos del array

# Multiplicación de matrices
producto_matriz = np.dot(matriz, np.array([1, 0, 1]))

# Mostrar los resultados
print("Array original:", array1)
print("Array sumado:", array_sum)
print("Array al cuadrado:", array_squared)
print("Suma de arrays:", array_add)
print("Media del array:", mean_array1)
print("Suma del array:", sum_array1)
print("Producto de la matriz:", producto_matriz)
```

```python
#Ejemplo pandas

# Crear un DataFrame a partir de un diccionario
data = {
    'Nombre': ['Ana', 'Luis', 'Carlos', 'María'],
    'Edad': [23, 30, 22, 35],
    'Ciudad': ['Madrid', 'Barcelona', 'Sevilla', 'Valencia'],
    'Puntuación': [85, 90, 78, 88]
}

df = pd.DataFrame(data)

# Mostrar el DataFrame completo
print("DataFrame completo:")
print(df)

# Acceder a una columna
print("\nColumna 'Nombre':")
print(df['Nombre'])

# Filtrar filas (por ejemplo, personas con Edad mayor a 25)
filtro_edad = df[df['Edad'] > 25]
print("\nFiltrar personas con edad mayor a 25:")
print(filtro_edad)

# Agregar una nueva columna calculada (Ej. Aumentar la Puntuación en un 10%)
df['Puntuación ajustada'] = df['Puntuación'] * 1.1
print("\nDataFrame con columna de Puntuación ajustada:")
print(df)

# Calcular estadísticas básicas (media de la edad y puntuación)
media_edad = df['Edad'].mean()
media_puntuacion = df['Puntuación'].mean()

print(f"\nMedia de la Edad: {media_edad}")
print(f"Media de la Puntuación: {media_puntuacion}")

# Guardar el DataFrame en un archivo CSV
df.to_csv('personas.csv', index=False)
print("\nDataFrame guardado como 'personas.csv'.")
```

```python
#Ejemplo scikit-learn

# Importar librerías
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Cargar el dataset Iris
iris = load_iris()
X = iris.data   # Las características (features)
y = iris.target # Las etiquetas (target)

# Dividir el dataset en conjunto de entrenamiento y prueba (70% entrenamiento, 30% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear el modelo de Regresión Logística
model = LogisticRegression(max_iter=200)

# Entrenar el modelo
model.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = model.predict(X_test)

# Evaluar el modelo
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

# Mostrar los resultados
print(f"Precisión del modelo: {accuracy}")
print("\nReporte de clasificación:")
print(report)
```
